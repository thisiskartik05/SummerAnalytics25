```python
import pandas as pd
train = pd.read_csv('Train_Data.csv')
test = pd.read_csv('Test_Data.csv')
```

```python
display(train.head())
display(test.head())
```

```python
print("Missing values in train data:")
print(train.isnull().sum())
print("\nMissing values in test data:")
print(test.isnull().sum())

```

```python
# missing values in numerical columns using the mean
numerical_cols = [ 'BMXBMI', 'LBXGLU', 'LBXGLT', 'LBXIN']
for col in numerical_cols:
    train[col].fillna(train[col].mean(), inplace=True)
    test[col].fillna(test[col].mean(), inplace=True)

```

```python
categorical_cols = ['RIAGENDR', 'PAQ605', 'DIQ010']
for col in categorical_cols:
    if train[col].isnull().any():
        train[col].fillna(train[col].mode()[0], inplace=True)
    if test[col].isnull().any():
        test[col].fillna(test[col].mode()[0], inplace=True)
```

```python


#Missing values after filling them TRAIN
print(train.isnull().sum())
#Missing values after filling them TEST
print(test.isnull().sum())

# Drop rows with missing age_group values from the training data
train.dropna(subset=['age_group'], inplace=True)

# Map 'age_group' to numerical values
train['age_group'] = train['age_group'].map({'Adult': 0, 'Senior': 1})

# Define features and drop the SEQN column
features = [col for col in train.columns if col not in ['SEQN', 'age_group']]

print("\nFeatures selected for modeling:")
print(features)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(train[features])

model = LogisticRegression(max_iter=1000)

model.fit(X_train_scaled, train['age_group'])
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(train[features], train['age_group'])
```

```python
# Scale the test data using the same scaler fitted on the training data
X_test_scaled = scaler.transform(test[features])

# Predict the age group on the scaled test data
predictions = model.predict(X_test_scaled)

# Display the predictions
display(predictions)
```

```python
# Create a submission DataFrame with SEQN and predictions
submission_df = pd.DataFrame({'SEQN': test['SEQN'], 'age_group': predictions})

# Save the submission DataFrame to a CSV file
submission_df.to_csv('submission.csv', index=False)

print("Submission file created successfully!")
```

```python
from sklearn.metrics import accuracy_score

# Predict the age group on the scaled training data
y_train_pred = model.predict(X_train_scaled)

accuracy = accuracy_score(train['age_group'], y_train_pred)

# accuracy score
print(f"Accuracy of the Logistic Regression model on the training data: {accuracy:.4f}")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# train the Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train_scaled, train['age_group'])

# train the Random Forest model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_scaled, train['age_group'])

# train the SVC model
svm_model = SVC(random_state=42)
svm_model.fit(X_train_scaled, train['age_group'])
```

```python
from sklearn.metrics import accuracy_score

# accuracy for Decision Tree
dt_train_pred = dt_model.predict(X_train_scaled)
dt_accuracy = accuracy_score(train['age_group'], dt_train_pred)
print(f"Accuracy of the Decision Tree model on the training data: {dt_accuracy:.4f}")

# accuracy for Random Forest
rf_train_pred = rf_model.predict(X_train_scaled)
rf_accuracy = accuracy_score(train['age_group'], rf_train_pred)
print(f"Accuracy of the Random Forest model on the training data: {rf_accuracy:.4f}")

# accuracy for Support Vector Machine
svm_train_pred = svm_model.predict(X_train_scaled)
svm_accuracy = accuracy_score(train['age_group'], svm_train_pred)
print(f"Accuracy of the Support Vector Machine model on the training data: {svm_accuracy:.4f}")
```

```python
accuracy_scores = {
    'Logistic Regression': accuracy,
    'Decision Tree': dt_accuracy,
    'Random Forest': rf_accuracy,
    'Support Vector Machine': svm_accuracy
}

accuracy_df = pd.DataFrame.from_dict(accuracy_scores, orient='index', columns=['Accuracy'])

# Sort the DataFrame by accuracy in descending order
accuracy_df = accuracy_df.sort_values(by='Accuracy', ascending=False)

# accuracy comparison
display(accuracy_df)
```

```python
# Use Logistic Regression model to predict on the scaled test data
lr_predictions = model.predict(X_test_scaled)
# Create and save the submission DataFrame for Logistic Regression
lr_submission_df = pd.DataFrame({'SEQN': test['SEQN'], 'age_group': lr_predictions})
lr_submission_df.to_csv('logistic_regression_submission.csv', index=False)
print("Logistic Regression submission file created successfully!")

# Use Decision Tree model to predict on the scaled test data
dt_predictions = dt_model.predict(X_test_scaled)
# Create and save the submission DataFrame for Decision Tree
dt_submission_df = pd.DataFrame({'SEQN': test['SEQN'], 'age_group': dt_predictions})
dt_submission_df.to_csv('decision_tree_submission.csv', index=False)
print("Decision Tree submission file created successfully!")

# Use Random Forest model to predict on the scaled test data
rf_predictions = rf_model.predict(X_test_scaled)
# Create and save the submission DataFrame for Random Forest
rf_submission_df = pd.DataFrame({'SEQN': test['SEQN'], 'age_group': rf_predictions})
rf_submission_df.to_csv('random_forest_submission.csv', index=False)
print("Random Forest submission file created successfully!")

# Use Support Vector Machine model to predict on the scaled test data
svm_predictions = svm_model.predict(X_test_scaled)
# Create and save the submission DataFrame for Support Vector Machine
svm_submission_df = pd.DataFrame({'SEQN': test['SEQN'], 'age_group': svm_predictions})
svm_submission_df.to_csv('svm_submission.csv', index=False)
print("Support Vector Machine submission file created successfully!")
```

```python
# Create a pandas DataFrame from the dictionary
updated_accuracy_df = pd.DataFrame.from_dict(accuracy_scores, orient='index', columns=['Accuracy'])

# Sort the DataFrame by accuracy in descending order
updated_accuracy_df = updated_accuracy_df.sort_values(by='Accuracy', ascending=False)

# Display the updated accuracy comparison
display(updated_accuracy_df)

# Identify the best performing model
best_model_name_updated = updated_accuracy_df.index[0]
best_accuracy_updated = updated_accuracy_df.iloc[0, 0]

print(f"\nBased on the accuracy on the training data, the best performing model is: {best_model_name_updated} with an accuracy of {best_accuracy_updated:.4f}")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np

# Split the training data to find the optimal ccp_alpha for pruning
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
    X_train_scaled, train['age_group'], test_size=0.2, random_state=42
)

# Find optimal ccp_alpha
path = dt_model.cost_complexity_pruning_path(X_train_split, y_train_split)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

# Train Decision Tree models with different ccp_alpha values
clfs = []
for ccp_alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha)
    clf.fit(X_train_split, y_train_split)
    clfs.append(clf)

# Evaluate accuracy on the validation set for each model
acc_scores = [accuracy_score(y_val_split, clf.predict(X_val_split)) for clf in clfs]

# Find the ccp_alpha that maximizes validation accuracy
optimal_ccp_alpha_index = np.argmax(acc_scores)
optimal_ccp_alpha = ccp_alphas[optimal_ccp_alpha_index]

print(f"Optimal ccp_alpha for pruning: {optimal_ccp_alpha:.4f}")

# Train the pruned Decision Tree model with the optimal ccp_alpha
pruned_dt_model = DecisionTreeClassifier(random_state=42, ccp_alpha=optimal_ccp_alpha)
pruned_dt_model.fit(X_train_scaled, train['age_group'])

# Evaluate the pruned Decision Tree model on the training data
pruned_dt_train_pred = pruned_dt_model.predict(X_train_scaled)
pruned_dt_accuracy = accuracy_score(train['age_group'], pruned_dt_train_pred)
print(f"Accuracy of the Pruned Decision Tree model on the training data: {pruned_dt_accuracy:.4f}")

# Add the pruned Decision Tree accuracy to the accuracy_scores dictionary
accuracy_scores['Pruned Decision Tree'] = pruned_dt_accuracy
```

```python
# Create a pandas DataFrame from the updated dictionary
updated_accuracy_df = pd.DataFrame.from_dict(accuracy_scores, orient='index', columns=['Accuracy'])

# Sort the DataFrame by accuracy in descending order
updated_accuracy_df = updated_accuracy_df.sort_values(by='Accuracy', ascending=False)

# Display the updated accuracy comparison
display(updated_accuracy_df)

# Identify the best performing model
best_model_name_updated = updated_accuracy_df.index[0]
best_accuracy_updated = updated_accuracy_df.iloc[0, 0]

print(f"\nBased on the accuracy on the training data, the best performing model is: {best_model_name_updated} with an accuracy of {best_accuracy_updated:.4f}")
```

```python
# Use Pruned Decision Tree model to predict on the scaled test data
pruned_dt_predictions = pruned_dt_model.predict(X_test_scaled)
# Create and save the submission DataFrame for Pruned Decision Tree
pruned_dt_submission_df = pd.DataFrame({'SEQN': test['SEQN'], 'age_group': pruned_dt_predictions})
pruned_dt_submission_df.to_csv('pruned_decision_tree_submission.csv', index=False)
```
